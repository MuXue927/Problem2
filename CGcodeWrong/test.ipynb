```python
import os
import pandas as pd

file_loc = os.path.join('../', 'datasets', 'multiple-periods', 'small')
dataset_name = 'dataset_1'
file_loc = os.path.join(file_loc, dataset_name)

file_name = 'warehouse_production.csv'
file = os.path.join(file_loc, file_name)
```

```python
df = pd.read_csv(file, header=0)
df.head()
```

```python
sku_prod_each_day = \
    df.set_index(['plant_code', 'product_code', 'produce_date'])['volume'].to_dict()
print(sku_prod_each_day)
```
